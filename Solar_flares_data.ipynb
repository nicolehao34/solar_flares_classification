# Processing Data


```python
# turn the txt into dataframe

# cut out part of the txt we want
df = pd.read_csv('solar_flares_data', on_bad_lines='skip')
df_ex = df.iloc[3:121209].copy()

df_ex.columns
```

```python
# split into columns


df_ex = df_ex['HESSI Flare List  (generated  7-Oct-2020 13:36)'].str.split(expand=True)

df_ex.head()
```

```python
# relabel columns
colnames=['Flare', 'Start', 'time', 'Peak', 'End','Dur (s)','Peak (c/s)','Total Counts', 'Energy (KeV)', 'X Pos (asec)', 'Y Pos (asec)', 'Radial (asec)', 'AR', 'Flag1','Flag2','Flag3','Flag4','Flag5','Flag6','Flag7','Flag8','Flag9','Flag10','Flag11']
df_ex.columns = colnames
df_ex
```

```python
# combine col 1 and col 2, separate by space
df_ex['Start'] = pd.to_datetime(df_ex['Start'])
df_ex["time JD"] = df_ex["Start"].astype(str) + ' ' + df_ex["time"].astype(str)
list1 = df_ex["time JD"].tolist()
t1 = Time(list1, scale = 'utc')

time_JD = pd.Series(t1.jd)
df_ex['time JD'] = time_JD.values

# repeat the process w list of peak
df_ex["Peak JD"] = df_ex["Start"].astype(str) + ' ' + df_ex["Peak"].astype(str)
list2 = df_ex["Peak JD"].tolist()
t2 = Time(list2,format = 'iso', scale = 'utc')

Peak_JD = pd.Series(t2.jd)
df_ex['Peak JD'] = Peak_JD.values

# list of end
df_ex["End JD"] = df_ex["Start"].astype(str) + ' ' + df_ex["End"].astype(str)
list3 = df_ex["End JD"].tolist()
t3 = Time(list3,format = 'iso', scale = 'utc')

End_JD = pd.Series(t3.jd)
df_ex['End JD'] = End_JD.values

# df_ex.drop(['JD time'], axis = 1, inplace = True)

df_ex
```

```python
# combine flag columns

index1 = df_ex.columns.get_loc("Flag1")
index2 = df_ex.columns.get_loc("Flag11")

df_ex['Flags'] = df_ex[df_ex.columns[index1:index2+1]].apply(lambda x: ', '.join(x.dropna().astype(str)), axis=1)

# remove 11 flag columns
df_ex.drop(['Flag1','Flag2','Flag3','Flag4','Flag5','Flag6','Flag7','Flag8','Flag9','Flag10','Flag11'], axis=1, inplace=True)    

df_ex
```

```python
# remove rows that have zeroes for both x-pos and y-pos

df_ex_1 = df_ex[df_ex['X Pos (asec)'] != 0]
df_ex_2 = df_ex_1[df_ex_1['Y Pos (asec)'] != 0]
df_ex_2
```
